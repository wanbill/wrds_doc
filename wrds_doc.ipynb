```python
import wrds
```

```python
db = wrds.Connection()
```

```python
sql_query_bs = """
SELECT 
    fyear,
    at AS "Total Asset",
    act AS "Current Assets",
    lct AS "Current Liabilities",
    che AS "Cash and Short-Term Investments",
    rect AS "Receivables",
    invt AS "Inventories",
    aco AS "Other Current Assets",
    ppent AS "Property, Plant, and Equipment (Net)",
    ivao AS "Investment and Advances (equity)",
    ivst AS "Investment Total (short-term)(Other Investments)",
    intan AS "Intangible Assets",
    ao AS "Other Assets",
    lt AS "Total Liabilities",
    dlc AS "Debt in Current Liabilities",
    ap AS "Account Payable",
    txp AS "Income Taxes Payable",
    lco AS "Other Current Liabilities",
    dltt AS "Long-term Debt",
    txditc AS "Deferred Taxes and Investment Tax Credit",
    lo AS "Other Liabilities",
    pstk AS "Preferred Stock",
    cstk AS "Common Stock",
    (at - lt) AS "Shareholders' Equity",
    teq AS "Stockholders' Equity Total",
    mib AS "Noncontrolling Interest",
    lse AS "Total Liabilities and Shareholders' Equity"
FROM comp.funda
WHERE fyear BETWEEN 1968 AND 2021
AND indfmt='INDL'
AND datafmt='STD'
AND popsrc='D'
AND consol='C'
AND curcd='USD'
AND fic='USA'
AND date_part('month', datadate) = 12 
AND date_part('day', datadate) = 31
AND prcc_f >= 1
AND at >= 1
AND ABS(at - lse) <= 1
--AND at = lt + teq
--  not sure if this correctAND (at = lt + teq) 
--AND fyear BETWEEN 1968 AND 2022
"""
bs_tb = db.raw_sql(sql_query_bs)
```
